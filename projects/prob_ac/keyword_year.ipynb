```python
import pandas as pd
```

```python
train_df = pd.read_csv("data/train_data.csv")
test_df = pd.read_csv("data/test_data.csv")
all_df = pd.concat([train_df, test_df], axis=0).reset_index(drop=True)
```

```python
from projects.prob_ac.custom.feature_generators import KeywordsLengthFeature, KeywordsAverageYear, KeywordsAverageCount, KeywordsFirstYear
from lilac.core.utils import plot_numeric_feature_hist_for_cls
```

```python
gens = [KeywordsLengthFeature(), KeywordsAverageYear(), KeywordsAverageCount(), KeywordsFirstYear()] 

for gen in gens:
    gen.fit(all_df)
    train_feat = gen.transform(train_df)
    test_feat = gen.transform(test_df)
    train_df=pd.concat([train_df,train_feat],axis=1)
    test_df=pd.concat([test_df,test_feat],axis=1)

train_df
```
