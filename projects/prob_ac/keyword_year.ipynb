```python
import pandas as pd
```

```python
train_df = pd.read_csv("data/train_data.csv")
test_df = pd.read_csv("data/test_data.csv")
all_df = pd.concat([train_df, test_df], axis=0).reset_index(drop=True)
```

```python
from projects.prob_ac.custom.feature_generators import KeywordsLengthFeature, KeywordsAverageYear, KeywordsAverageCount
```

```python
gens = [KeywordsLengthFeature(), KeywordsAverageYear(), KeywordsAverageCount()] 

for gen in gens:
    gen.fit(all_df)
    train_feat = gen.transform(train_df)
    test_feat = gen.transform(test_df)
    train_df=pd.concat([train_df,train_feat],axis=1)
    test_df=pd.concat([test_df,train_feat],axis=1)

train_df
```
