```python
import pandas as pd
```

```python
train_data=pd.read_csv("data/train_data.csv")
train_data
```

```python
sub_data=pd.read_csv("data/submission.csv")
sub_data
```

```python
id_list = train_data["id"]
dropped=train_data.drop("id",axis= 1)
transposed=dropped.transpose()
transposed.columns=id_list
transposed.reset_index(inplace=True)
transposed=transposed.rename(columns={"index":"id"})
```

```python
from sklearn.metrics import mean_squared_log_error
import numpy as np

transposed["2019-11-01-pred"]=np.exp((np.log(transposed["2018-11-01"]) + np.log(transposed["2019-10-01"]))/2)
transposed["2019-11-01-pred"]=(transposed["2018-11-01"] + transposed["2019-10-01"])/2
np.sqrt(mean_squared_log_error(transposed["2019-11-01"], transposed["2019-11-01-pred"]))
```

```python
transposed["y"]=np.exp((np.log(transposed["2018-12-01"]) + np.log(transposed["2019-11-01"]))/2)
```

```python
transposed[["id","y"]].to_csv("data/subs/baseline2.csv",index=False)
```
